# CA06 KNN Movie Recommendation

Loading in dataset

```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
import io 
import requests
import matplotlib.pyplot as mp
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
```

```python
url="https://github.com/ArinB/CA05-kNN/raw/master/movies_recommendation_data.csv"
s=requests.get(url).content
movies=pd.read_csv(io.StringIO(s.decode('utf-8')))
```

Creating Features

```python
#Drop Label as it is useless 
movies= movies.drop('Movie ID', axis=1)
```

```python
movies['IMDB Rating'] = movies['IMDB Rating'].astype(int)#Changing IMDB Rating from float to integer
```

```python
x = movies.drop('Movie Name', axis=1) #Independent
y = movies['Movie Name'] #Dependent 
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
```

# KNN Classifier

```python
k = 5
knn = KNeighborsClassifier(n_neighbors=k).fit(x_train,y_train)#Passing parameters through the KNN model
```

```python
y_pred=knn.predict(x_test)
```

```python
confusion=confusion_matrix(y_test,y_pred)
accuracy=accuracy_score(y_test,y_pred)
```
